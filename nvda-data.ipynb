```python
#Wacc = Financial Leverage x Cost of Debt + (1 - Financial Leverage) x Cost of Equity
mc = nvda.info["marketCap"]
debt = nvda.info["totalDebt"]
mv = mc + debt
treasury_yield10 = yf.Ticker("^TNX")
risk_free_rate = round(treasury_yield10.history()["Close"].iloc[-1]/100,2)
sp500_exp_ret = 0.10
beta = nvda.info["beta"]
cost_of_equity = round(risk_free_rate+beta*(sp500_exp_ret-risk_free_rate),2)
balance_sheet = nvda.balance_sheet
weight_of_equity = balance_sheet.loc["Stockholders Equity"].iloc[0]/balance_sheet.loc["Total Assets"].iloc[0]
stock_fin = nvda.financials
cost_of_debt = round(stock_fin.loc["Interest Expense"].iloc[0] *-1 / balance_sheet.loc["Current Liabilities"].iloc[0],2)
weight_of_debt = round(balance_sheet.loc["Current Liabilities"].iloc[0]/ balance_sheet.loc["Total Assets"].iloc[0],2)
tax_rate = round(stock_fin.loc["Tax Provision"].iloc[0] / stock_fin.loc["Pretax Income"].iloc[0],2)

wacc = round((weight_of_equity * cost_of_equity) + ((weight_of_debt * cost_of_debt ) * (1-tax_rate)),2)
#wacc = mc/1
wacc
```
